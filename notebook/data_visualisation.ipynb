```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/clean.csv")
df
```

```python
df = df.drop("Unnamed: 0" , axis=1)
```

```python
df
```

```python
class DataVisualisation:
    
    def __init__(self,df):
        self.df = df
    
    def info(self): 

        #print("is the dataset balanced or not : ")
        
        df["Loan_Status"].value_counts().plot.pie()

        #print("Found outliers : ")
        df.plot(kind="box" , subplots=True , figsize=(15,15) , title="Found Outliers")
        
           #

        

        #print("Analyze the data column by column")
        for column in df:
            df[column].value_counts().sort_index(axis=0).plot(kind='bar')
            plt.title(column)
            plt.show()
```

```python
test_obj = DataVisualisation(df)
test_obj.info()
```
